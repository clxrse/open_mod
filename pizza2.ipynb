```python
import matplotlib.pyplot as plt
from highlight_text import fig_text
from mplsoccer import PyPizza, FontManager
```

```python
font_normal = FontManager(("https://github.com/clxrse/FONTS/blob/main/Oswald-Regular.ttf?raw=true"))
font_italic = FontManager(("https://github.com/clxrse/FONTS/blob/main/Oswald-RegularItalic.ttf?raw=true"))
font_bold = FontManager(("https://github.com/clxrse/FONTS/blob/main/Oswald-Bold.ttf?raw=true"))
```

```python
# parameter and values list
# The values are taken from the excellent fbref website (supplied by StatsBomb)
params = [
    "Non-Penalty Goals", "npxG", "Assists", "xA",
    "Open Play\nShot Creating Actions", "\nPenalty Area\nTouches",
    "Progressive Passes", "Progressive Carries", "Successful Dribbles",
    "\nProgressive\nPasses Rec.", "pAdj Pressures", "Aerials Won"
]
values = [85, 67, 99, 99, 99, 48, 99, 94, 90, 92, 81, 79]    # for Robert Lewandowski
values_2 = [83, 99, 87, 98, 99, 96, 99, 98, 99, 95, 49, 32]  # for Mohamed Salah

# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    background_color="#EBEBE9",     # background color
    straight_line_color="#222222",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=1,               # linewidth of last circle
    last_circle_color="#222222",    # color of last circle
    other_circle_ls="-.",           # linestyle for other circles
    other_circle_lw=1               # linewidth for other circles
)

# plot pizza
fig, ax = baker.make_pizza(
    values,                     # list of values
    compare_values=values_2,    # comparison values
    figsize=(8, 8),             # adjust figsize according to your need
    kwargs_slices=dict(
        facecolor="#1A78CF", edgecolor="#222222",
        zorder=2, linewidth=1
    ),                          # values to be used when plotting slices
    kwargs_compare=dict(
        facecolor="#FF9300", edgecolor="#222222",
        zorder=2, linewidth=1,
    ),
    kwargs_params=dict(
        color="#000000", fontsize=12,
        fontproperties=font_normal.prop, va="center"
    ),                          # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=12,
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    ),                          # values to be used when adding parameter-values labels
    kwargs_compare_values=dict(
        color="#000000", fontsize=12, fontproperties=font_normal.prop, zorder=3,
        bbox=dict(edgecolor="#000000", facecolor="#FF9300", boxstyle="round,pad=0.2", lw=1)
    ),                          # values to be used when adding parameter-values labels
)

# add title
fig_text(
    0.515, 0.99, "<Ruslan Malinovskyi> vs <Neymar Jnior>", size=17, fig=fig,
    highlight_textprops=[{"color": '#1A78CF'}, {"color": '#EE8900'}],
    ha="center", fontproperties=font_bold.prop, color="#000000"
)

# add subtitle
fig.text(
    0.515, 0.942,
    "Percentile Rank vs Positional Peers | Season 2020-21",
    size=15,
    ha="center", fontproperties=font_bold.prop, color="#000000"
)

# add credits
CREDIT_1 = "data: statsbomb viz fbref"
CREDIT_2 = "inspired by: @Worville, @FootballSlices, @somazerofc & @Soumyaj15209314"
CREDIT_3 = "@keratinised"
fig.text(
    0.99, 0.005, f"{CREDIT_1}\n{CREDIT_2}\n{CREDIT_3}", size=9,
    fontproperties=font_italic.prop, color="#000000",
    ha="right"
)
from mplsoccer import add_image
from urllib.request import urlopen
from PIL import Image
right_arrow_url = 'https://media.atalanta.it/wp-content/uploads/2018/07/Giocatore-13.png'
right_arrow = Image.open(urlopen(right_arrow_url))
right_arrow_url2 = 'https://raw.githubusercontent.com/clxrse/FONTS/main/alt-card_neymar.png'
right_arrow2 = Image.open(urlopen(right_arrow_url2))
ax_right_arrow = add_image(right_arrow,fig,left= 0.1,bottom=0.85,width=(0.1),alpha=1)
ax_right_arrow2 = add_image(right_arrow2,fig,left= 0.8,bottom=0.85,width=(0.15),alpha=1)


plt.show()
plt.savefig("Pizza.png",dpi=800)
```
